<h3>Term Paper & Presentation</h3>

<table>
  <thead>
    <tr>
      <th rowspan="2" style="text-align:left;">Dimension</th>
      <th colspan="2" style="text-align:center;">Excellent (A+/A)</th>
      <th colspan="2" style="text-align:center;">Moderate (C- to B+)</th>
      <th colspan="2" style="text-align:center;">Inadequate (D/F)</th>
    </tr>
    <tr>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><b>Thesis & Argument</b></td>
      <td>Clear thesis connected to course frameworks; arguments flow logically; claims scoped appropriately.</td>
      <td>Original thesis that advances or critiques frameworks; anticipates counterarguments.</td>
      <td>Thesis present but descriptive; logical gaps.</td>
      <td>Unclear thesis; weak organization.</td>
      <td>Vague or shifting claims.</td>
      <td>No discernible argument.</td>
    </tr>
    <tr>
      <td><b>Engagement with Literature</b></td>
      <td>Cites core readings accurately; integrates at least 6–8 relevant sources.</td>
      <td>Synthesizes primary literature; contrasts methods/results; includes 10+ high-quality sources.</td>
      <td>Uses some literature; limited synthesis.</td>
      <td>Relies on tertiary summaries; minimal synthesis.</td>
      <td>Few or outdated sources.</td>
      <td>No meaningful engagement with literature.</td>
    </tr>
    <tr>
      <td><b>Analysis & Evidence</b></td>
      <td>Includes figures/tables produced by the student; correct methods; limitations noted.</td>
      <td>Employs advanced analysis (e.g., IAM scenario critique, sensitivity); triangulates evidence.</td>
      <td>Some analysis; limited rigor or transparency.</td>
      <td>Descriptive; assertions lack support.</td>
      <td>Misinterprets evidence; errors in analysis.</td>
      <td>No analysis or fabricated evidence.</td>
    </tr>
    <tr>
      <td><b>Writing & Presentation</b></td>
      <td>Clear, concise, well-structured; proper citations; meets page limits.</td>
      <td>Professional polish akin to policy brief or conference paper; consistent style.</td>
      <td>Readable but uneven organization.</td>
      <td>Disorganized; style inconsistent.</td>
      <td>Hard to follow; many errors.</td>
      <td>Unacceptable academic writing.</td>
    </tr>
  </tbody>
</table>

      <th colspan="2" style="text-align:center;">Excellent (A+/A)</th>
      <th colspan="2" style="text-align:center;">Moderate (C- to B+)</th>
      <th colspan="2" style="text-align:center;">Inadequate (D/F)</th>
    </tr>
    <tr>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><b>Conceptual Understanding</b></td>
      <td>Accurate application of climate/WEF concepts; correct formulae and definitions with justification.</td>
      <td>Critically evaluates conceptual trade-offs and boundary conditions; situates concepts within frameworks.</td>
      <td>Mostly correct but with minor conceptual gaps.</td>
      <td>Partial understanding; key definitions misapplied.</td>
      <td>Major misconceptions.</td>
      <td>No understanding shown.</td>
    </tr>
    <tr>
      <td><b>Application to Data</b></td>
      <td>Uses provided datasets/figures; includes hand-drawn sketches or calculations created by student.</td>
      <td>Designs/extends analysis (e.g., scenario variant, robustness check) and justifies methods.</td>
      <td>Attempts data use; limited interpretation.</td>
      <td>Minimal data engagement; descriptive only.</td>
      <td>No data used.</td>
      <td>Irrelevant or erroneous data usage.</td>
    </tr>
    <tr>
      <td><b>Originality & Independence</b></td>
      <td>Own voice evident; personalized examples; clear attribution; passes spot oral checks.</td>
      <td>Novel framing or critique; integrates external sources; passes oral defense on methods/choices.</td>
      <td>Some originality; heavy paraphrasing.</td>
      <td>Generic responses; limited independence.</td>
      <td>Signals of AI-generated content; lacks authorship markers.</td>
      <td>Copied, AI-produced, or uncredited material.</td>
    </tr>
    <tr>
      <td><b>Clarity & Format</b></td>
      <td>Well-structured; crisp figures with captions; reproducible steps if code used.</td>
      <td>Professional formatting; consistent style; appendix for methods if advanced.</td>
      <td>Mostly clear; some formatting issues.</td>
      <td>Organization issues; unclear visuals.</td>
      <td>Poor formatting; hard to read.</td>
      <td>Non-compliant submission.</td>
    </tr>
  </tbody>
</table>


<h3>Projects (Capstone/Research)</h3>

  <thead>

      <th rowspan="2" style="text-align:left;">Dimension</th>
      <th colspan="2" style="text-align:center;">Excellent (A+/A)</th>
      <th colspan="2" style="text-align:center;">Moderate (C- to B+)</th>
      <th colspan="2" style="text-align:center;">Inadequate (D/F)</th>
    </tr>
    <tr>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
      <th style="text-align:center;">Level 400</th>
      <th style="text-align:center;">Level 500</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><b>Problem Definition & Relevance</b></td>
      <td>Focused question tied to course goals and local/regional context; measurable scope.</td>
      <td>Research-grade framing aligned with active policy debates; explicit novelty.</td>
      <td>Relevant but broad; scope creep risks.</td>
      <td>Loose/unclear problem statement.</td>
      <td>Tangential to course aims.</td>
      <td>No clear problem.</td>
    </tr>
    <tr>
      <td><b>Methods & Modeling</b></td>
      <td>Appropriate tool choice (e.g., EN-ROADS/MESSAGE); sound assumptions; transparent workflow.</td>
      <td>Extends tools or triangulates models/datasets; sensitivity and uncertainty analysis.</td>
      <td>Applies tools but with gaps or limited justification.</td>
      <td>Superficial tool use; black-box results.</td>
      <td>Inappropriate or unjustified methods.</td>
      <td>No credible methodology.</td>
    </tr>
    <tr>
      <td><b>Governance Linkages</b></td>
      <td>Translates results into concrete policy options; trade-offs identified.</td>
      <td>Develops implementable pathways; discusses feasibility, equity, and capability considerations.</td>
      <td>Mentions policy links but shallow.</td>
      <td>Weak or generic policy translation.</td>
      <td>No clear policy implications.</td>
      <td>Irrelevant to governance.</td>
    </tr>
    <tr>
      <td><b>Teamwork & Communication</b></td>
      <td>Clear division of roles; coherent narrative; high-quality visuals authored by team.</td>
      <td>Professional delivery; effective Q&A; evidence of iteration from feedback.</td>
      <td>Functional but uneven contributions.</td>
      <td>Coordination problems; unclear narrative.</td>
      <td>Major collaboration issues.</td>
      <td>Disjoint or incomplete.</td>
    </tr>
  </tbody>
</table>
