## 📚 Core & Extended Readings

### Core Readings
- **“The Discovery of Global Warming”** by Spencer Weart ([online book])  
- **“The Britannica Guide to Climate Change: An Unbiased Guide to the Key Issue of Our Age”** (Introduction by Robert May)  
- **“The New Climate War: The Fight to Take Back Our Planet”** by Michael E. Mann (selected chapters)  
- **“Introduction to Climate Science”** by Andreas Schmittner ([online book])  

### Extended/Recommended Readings
- **“How to Avoid a Climate Disaster”** by Bill Gates (2021) – Technological & policy solutions  
- **“On the Move: Energy, Environment, and the Economy”** by Vaclav Smil (2013)  
- **“Energy and Civilization: A History”** by Vaclav Smil (2017)  
- **“Growth: From Microorganisms to Megacities”** by Vaclav Smil (2019)  
- **“No One Is Too Small to Make a Difference”** by Greta Thunberg (2019) – Collection of speeches  
- **“The Climate Book”** by Greta Thunberg (2022) – Essays by scientists & activists curated by Thunberg  

### Textbooks & Academic References
- **“Climate Change 2021: The Physical Science Basis”** – Intergovernmental Panel on Climate Change (IPCC AR6, Working Group I Report)  
- **“Climate Change 2022: Impacts, Adaptation and Vulnerability”** – IPCC AR6, Working Group II Report  
- **“Climate Change: The Science of Global Warming and Our Energy Future”** by Edmond A. Mathez & Jason E. Smerdon (2018)  
- **“Global Warming: Understanding the Forecast”** by David Archer (2011)  
- **“The Climate Crisis: An Introductory Guide to Climate Change”** by David Archer & Stefan Rahmstorf (2010)  
- **“Climate System Dynamics and Modelling”** by Hugues Goosse (2015) ([open access online])  
- **“Our Warming Planet: Topics in Climate Dynamics”** by Stuart Gaffin (Cambridge, 2012)  
- **“An Introduction to Climate Change: Lecture Notes”** by Andrew Dessler (commonly used in courses, multiple editions)  
- **“Climate Change and Society: Sociological Perspectives”** edited by Riley E. Dunlap & Robert J. Brulle (2015) – Links science to society and policy  
- **“Energy Transitions: Global and National Perspectives”** by Vaclav Smil (2017) – Energy perspective on climate  
- **“Principles of Planetary Climate”** by Raymond T. Pierrehumbert (2010) – Mathematical/physical foundation for climate dynamics  

### Classic Foundational Works
- **“Silent Spring”** by Rachel Carson (1962) – Early environmental warning  
- **“The Population Bomb”** by Paul R. Ehrlich (1968) – Debate on population and environmental stress  
- **“Limits to Growth”** by Donella Meadows et al. (1972; 30-year update in 2004) – Systems modeling of sustainability

## 🎬 Documentaries (by Theme)

### 1. Science & Environmental Impacts
- *Meltdown* (2021, Fredric Goldring) – Arctic melting  
- *Chasing Ice* (2012, James Balog) – Polar ice melt  
- *Antarctica: Ice and Sky* (2015, Luc Jacquet) – Field science in Antarctica  
- *Breaking Boundaries: The Science of Our Planet* (2021, David Attenborough) – Earth system tipping points  
- *Our Planet* (2019, Netflix, narrated by David Attenborough) – Ecosystems & biodiversity  
- *A Life on Our Planet* (2020, David Attenborough) – Witness statement on planetary decline  
- *More than Honey* (2013, Markus Imhoof) – Biodiversity & bees  

### 2. Politics, Economics & Skepticism
- *Merchants of Doubt* (2014, Robert Kenner) – Politics & creation of climate skepticism  
- *The Age of Consequences* (2016, Jared Scott) – Climate as driver of conflict & migration  
- *The Island President* (2011, Jon Shenk) – Small island nations & climate diplomacy  
- *Inconsistent Truth* (2012, Shayne Edwards) – Skepticism against climate change  
- *An Inconvenient Truth* (2006, Davis Guggenheim) – Landmark public awareness  
- *An Inconvenient Sequel: Truth to Power* (2017, Bonni Cohen & Jon Shenk) – Policy & activism follow-up  

### 3. Solutions & Activism
- *2040* (2019, Damon Gameau) – Future solutions & positive vision  
- *Before the Flood* (2016, Fisher Stevens, Leonardo DiCaprio) – Global impacts & policy solutions  
- *Kiss the Ground* (2020, Josh & Rebecca Tickell) – Soil regeneration & carbon farming  
- *Ice on Fire* (2019, Leila Conners, produced by Leonardo DiCaprio) – Climate innovations  
- *This Changes Everything* (2015, Avi Lewis, based on Naomi Klein’s book) – Climate justice & economic systems  
- *How to Let Go of the World and Love All the Things Climate Can’t Change* (2016, Josh Fox) – Global activism  
- *Years of Living Dangerously* (2014–2016, Showtime Series) – Journalistic exploration of climate solutions  
- *I Am Greta* (2020, Nathan Grossman) – Greta Thunberg’s activism journey  
- *The Human Element* (2018, James Balog) – Human stories in climate impacts  
- *Thank You for the Rain* (2017, Julia Dahr) – Farmer-turned-activist story in Kenya  
- *Revolution* (2012, Rob Stewart) – Global activism & resilience  

### 4. Food, Agriculture & Industry
- *Cowspiracy: The Sustainability Secret* (2014) – Livestock industry’s impact on climate  
- *Gasland Part II* (2016, Josh Fox) – Fracking & energy sector impacts  
- *The Great Warming* (2006, Michael Taylor) – Cultural/public impacts  

### 5. Future Scenarios
- *The Age of Stupid* (2009, Franny Armstrong) – Dramatized look at future if no action taken  
- *The 11th Hour* (2007, Leonardo DiCaprio, Nadia & Leila Conners) – Urgency for action  
- *Burning* (2021, Eva Orner) – Australia’s wildfires and climate link
